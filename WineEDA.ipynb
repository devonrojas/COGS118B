```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

redWineDF = pd.read_csv("redwinedata.csv", sep=";") # Load in red wines data
whiteWineDF = pd.read_csv("whitewinedata.csv", sep=";")  # Load in white wines data

bothWineDF = pd.concat([redWineDF, whiteWineDF]) # Combine DF into single DF
bothWineDF = bothWineDF.sample(frac=1).reset_index(drop=True) # Randomize the combined DF
```

### Intial Data
Our intial data starts with two data sets one related to red wines and one related to white wines.
We then combined both of these data sets into a single dataframe and randomized the entries.

The data set of red wines has 1599 rows and 12 columns. (1599 entries with 12 dimensions)

The data set of white wines has 4898 rows and 12 columns. (4989 entries with 12 dimensions)

The full data set has 6497 rows and 12 columns. (6497 entries with 12 dimensions)

##### The Data columns are:
1. fixed acidity
2. volatile acidity
3. citric acid
4. residual sugar
5. chlorides
6. free sulfur dioxide
7. total sulfur dioxide
8. density
9. pH
10. sulphates
11. alcohol
12. quality (score between 0 and 10)

```python
headers = ["Red Wines", "White Wines", "Both Wines"]
columnNames = ["fixed acidity", "volatile acidity", "residual sugar", "pH", "quality"]

for x in columnNames:
    plt.subplots(1)
    plt.title(x)
    plt.boxplot([redWineDF[x], whiteWineDF[x], bothWineDF[x]], labels=headers)
```

```python
columnNames = [["fixed acidity", "pH"], ["chlorides", "residual sugar"], ["residual sugar", "pH"]]

fig, axes = plt.subplots(len(columnNames), len(headers))

fig.set_figheight(18)
fig.set_figwidth(15)

rowAxes = 0
for x in columnNames:
    colAxes = 0
    
    axes[rowAxes, colAxes].set(xlabel= x[0], ylabel= x[1], title= "Red Wines")
    axes[rowAxes, colAxes].scatter(redWineDF[x[0]], redWineDF[x[1]])
    
    axes[rowAxes, colAxes + 1].set(xlabel= x[0], ylabel= x[1], title= "White Wines")
    axes[rowAxes, colAxes + 1].scatter(whiteWineDF[x[0]], whiteWineDF[x[1]])
    
    axes[rowAxes, colAxes + 2].set(xlabel= x[0], ylabel= x[1], title="Both Wines")
    axes[rowAxes, colAxes + 2].scatter(bothWineDF[x[0]], bothWineDF[x[1]])
    
    rowAxes += 1
    
```

### Initial EDA
Our preliminary analysis shows that there should be enough difference between red wines and white wines that a machine learing algorithm is capable of identifying whether a wine is red or white given some parameters.
