```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

redWineDF = pd.read_csv("winequality-red.csv", sep=";") # Load in red wines data
whiteWineDF = pd.read_csv("winequality-white.csv", sep=";")  # Load in white wines data

bothWineDF = pd.concat([redWineDF, whiteWineDF]) # Combine DF into single DF
bothWineDF = bothWineDF.sample(frac=1).reset_index(drop=True) # Randomize the combined DF
```

### Intial Data
Our intial data starts with two data sets one related to red wines and one related to white wines.
We then combined both of these data sets into a single dataframe and randomized the entries.

The data set of red wines has 1599 rows and 12 columns. (1599 entries with 12 dimensions)

The data set of white wines has 4898 rows and 12 columns. (4989 entries with 12 dimensions)

The full data set has 6497 rows and 12 columns. (6497 entries with 12 dimensions)

##### The Data columns are:
1. fixed acidity (float)
2. volatile acidity (float)
3. citric acid (float)
4. residual sugar (float)
5. chlorides (float)
6. free sulfur dioxide (float)
7. total sulfur dioxide (float)
8. density (float)
9. pH (float)
10. sulphates (float)
11. alcohol (float)
12. quality (int)

```python
headers = ["Red Wines", "White Wines", "Both Wines"]
for x in bothWineDF.columns.values:
    plt.subplots(1)
    plt.title(x)
    plt.boxplot([redWineDF[x], whiteWineDF[x], bothWineDF[x]], labels=headers)
```

##### The boxplot above show that for a significant amount of the categories the data contains outliers. The lowest outliers are from quality and alcohol.

```python
plt.figure(figsize= (20, 16))
for i, col in enumerate(list(bothWineDF.columns.values)):
    plt.subplot(4, 4, i + 1)
    sns.histplot(bothWineDF[col], kde= True)
    plt.grid()
    plt.tight_layout()
```

##### The histogram plots lets us the distribution of our data and it becomes clear that the only normally distributed parameter is pH. The rest of the parameters seem to have an approximate normal distribution with a postive skew. 

```python
sns.pairplot(data= bothWineDF, kind="scatter")
```

##### The plots above graph all possible pairs of parameters in our dataset and the result is that there is relatively low correlation between any two parameters.
